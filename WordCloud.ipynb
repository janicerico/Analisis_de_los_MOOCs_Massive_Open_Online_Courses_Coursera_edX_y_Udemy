## Generación de un WordCloud

```python
# Se importan las librerías a utilizar.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud
```

```python
# Se utiliza el dataset de Coursera_courses para usar las palabras de la columna 'name'.

df_texto = pd.read_csv ('./Datasets/Coursera_courses.csv', usecols=['name'])
```

```python
# Se convierte a string.

texto = str(df_texto)
```

```python
# Se abre la imagen con la forma que tendrá el WordCloud y se convierte en un array.

mask = np.array(Image.open('./WordCloud/bombillo.png'))

mask[mask == 1] = 255
```

```python
# Se genera el WordCloud.

word_cloud = WordCloud(background_color='white', contour_width=1, contour_color='grey', max_words=1000, min_font_size=5, collocation_threshold=10, mask=mask).generate(texto)

# Se guarda la imagen generada.

word_cloud.to_file('./WordCloud/wordcloud.png')

plt.figure(figsize=(10,8))
plt.imshow(word_cloud)
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```
